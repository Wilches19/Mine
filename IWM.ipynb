# Doing a Portafolio of Quantfury Stocks

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
symbols = ["IWM", "DKNG","KR","RIVN","PLUG","JD","BABA","BOIL","CHPT"]
# Lista para almacenar los precios de oferta
bid_prices = []
# Itera sobre cada símbolo
for symbol in symbols:
    # Utiliza la función 'Ticker' para obtener información en tiempo real
    stock_ticker = yf.Ticker(symbol)
    
    # Obtiene el precio de oferta (bid) directamente
    bid_price = stock_ticker.info['bid']
    
    # Añade el precio de oferta a la lista
    bid_prices.append(bid_price)

tabla = pd.DataFrame(index=())
tabla.index.name = 'Acciones'
tabla['Ticket'] = symbols
tabla['Precio de compra de activo'] = [170.73, 26.56,44.38,17.28,3.96,27.73,76.80,22.81,2.23] 
tabla['Valor'] = bid_prices
tabla['Cantidad del Activo'] = [1, 1,1,2,5,2,1,3,5] 
tabla['Valor Inicial'] = tabla['Precio de compra de activo'] * tabla['Cantidad del Activo']
V0=tabla['Valor Inicial'].sum()
tabla['Valor Hoy'] = tabla['Valor'] * tabla['Cantidad del Activo']
V1=tabla['Valor Hoy'].sum()
tabla['Rate %'] = ((tabla['Valor Hoy'] / tabla['Valor Inicial'] - 1) * 100)

# Mostrar DataFrame con formato condicional

# Graficar
tabla.plot.pie(y='Valor Hoy', labels=tabla['Ticket'], autopct='%1.2f%%') #puede ser 
def highlight_color(val):
    color = 'red' if val < 0 else 'green'
    return f'color: {color}'

# Aplicar formato condicional a la columna 'Rate'
tabla = tabla.style.applymap(highlight_color, subset=['Rate %'])
plt.axis('equal')  # Equal aspect ratio asegura que el gráfico sea un círculo en lugar de una elipse
plt.title('Distribución del Valor de Hoy por Acción')
plt.ylabel('')  # Elimina el nombre del eje y
plt.show()
tabla #impresion del DataFrame
```

```python
print(V0,"Es la cantidad invertida y", V1.round(2), "Es la cantidad en Activos, en general los beneficios son ", V1-V0,"en dolares y en porcentaje es %", ((V1/V0)-1)*100)
```

# Quantfury Crypto

```python
symbols = ["BTC-USD","MATIC-USD","ETH-USD","ETH-USD"]
# Lista para almacenar los precios de oferta
bid_prices = []
# Itera sobre cada símbolo
for symbol in symbols:
    # Utiliza la función 'Ticker' para obtener información en tiempo real
    stock_ticker = yf.Ticker(symbol)
    
    # Obtiene el precio de oferta (bid) directamente
    bid_price = stock_ticker.info['open']
    
    # Añade el precio de oferta a la lista
    bid_prices.append(bid_price)

tabla = pd.DataFrame(index=())
tabla.index.name = 'Acciones'
tabla['Ticket'] = symbols
tabla['Precio de compra de activo'] = [26437.36, 0.5598,1629.96,1717.37] 
tabla['Valor'] = bid_prices
tabla['Cantidad del Activo'] = [10, 10, 30, 10] 
tabla["Fraccion de activos"]=tabla['Cantidad del Activo']/tabla['Precio de compra de activo'] 
tabla['Valor Final']= tabla['Valor'] * tabla['Fraccion de activos']
V2=tabla['Cantidad del Activo'].sum()
tabla['Rate $'] = tabla['Valor Final'] - tabla['Cantidad del Activo']
V3=tabla['Valor Final'].sum()
tabla['Rate %'] = ((tabla['Valor Final'] / tabla['Cantidad del Activo'] - 1) * 100)

# Mostrar DataFrame con formato condicional

# Graficar
tabla.plot.pie(y='Valor Final', labels=tabla['Ticket'], autopct='%1.2f%%') #puede ser y=Cantidad del Activo o y=Valor Final
def highlight_color(val):
    color = 'red' if val < 0 else 'green'
    return f'color: {color}'

# Aplicar formato condicional a la columna 'Rate'
tabla = tabla.style.applymap(highlight_color, subset=['Rate %'])
plt.axis('equal')  # Equal aspect ratio asegura que el gráfico sea un círculo en lugar de una elipse
plt.title('Distribución del Valor de Hoy por Acción')
plt.ylabel('')  # Elimina el nombre del eje y
plt.show()
tabla #impresion del DataFrame
```

```python
print(V2,"Es la cantidad invertida y", V3.round(2), "es la cantidad en Activos, los beneficios son ", V3-V2,"en dolares y en porcentaje es %", ((V3/V2)-1)*100)
```
